```python
import pandas as pd
```

```python
df = pd.read_csv('/content/world_population.csv')
df.head()
```

```python
#to see the specific columns
df[['Rank', 'Capital']]
```

```python
#to see the specific row with string value
specific_countries =['Bangladesh', 'Brazil']
df[df['Country'].isin(specific_countries)]
```

```python
df[df['Country'].str.contains('United')]
```

```python
#copy the dataframe in another variable
#set a coulmn as new index
df2 = df.set_index('Country')
df2
```

```python
#use filter for showing specific columns
#axis = 1 , for searching horizontally
df2.filter(items =['Continent', 'CCA3'], axis = 1)
```

```python
#axis = 0 , for searching vertically
df2.filter(items =['Zimbabwe'], axis = 0)
```

```python
#treat the first column as integer numberring and show the value of that exact row
df2.iloc[3]
```

```python
df2.loc['United States']
```

```python
#to see by order and with conditions
df[df['Rank'] < 10].sort_values(by = 'Rank', ascending = False)
```

```python
#sort by any column in dataframe
df.sort_values (by = 'Country', ascending = False)
```
